```python
#Recommendation Model: Some of the movies hadn’t been watched and therefore, are not rated by the users. 
#Netflix would like to take this as an opportunity and build a machine learning recommendation algorithm which provides the ratings for each of the users.
```

```python
movies_df=pd.read_csv("Amazon - Movies and TV Ratings.csv")
```

creating association matrix

```python
movies_df.drop(['user_id'],axis=1,inplace=True)
```

```python
movies_df.columns
```

```python
learningMatrix =movies_df
learningMatrix.fillna(0,inplace=True)
learningMatrix
```

```python
#Divide the data into training and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data_matrix, test_data_matrix=train_test_split(learningMatrix, test_size=0.25 )
```

```python
#Build a recommendation model on training data
```

apply Cosine Similarity Formula on Association Matrix

```python
from scipy.spatial.distance import cosine
from sklearn.metrics import pairwise_distances
user_similarity =1-pairwise_distances(train_data_matrix, metric="cosine" )
np.fill_diagonal(user_similarity, 0 ) 
ratings_matrix = pd.DataFrame(user_similarity )
ratings_matrix
```

```python
#Make predictions on the test data
```

```python
mean_user_rating = train_data_matrix.mean(axis=1)[:, np.newaxis] 
mean_user_rating
```

```python
ratings_diff = (train_data_matrix - mean_user_rating)
ratings_diff
```

```python
# User based collaborative filtering model:
user_pred = mean_user_rating + user_similarity.dot(ratings_diff) / np.array([np.abs(user_similarity).sum(axis=1)]).T
user_pred
```

```python
#item based collaborative filtering 
movie_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
movie_pred = train_data_matrix.dot(movie_similarity)/ np.array([np.abs(movie_similarity).sum(axis=1)])
            
movie_pred
```
